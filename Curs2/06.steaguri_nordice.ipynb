```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
SCALE = 1 * 100

WIDTH = 15 * SCALE
HEIGHT = 9 * SCALE
```

```python
steag_arr = np.ones((HEIGHT, WIDTH,  4), dtype=np.uint8)
```

```python
# ALPHA
steag_arr[:,:,-1] = 255
```

```python
BACKGROUND_COLOR = np.array([0, 43, 127])
CROSS_COLOR = np.array([252, 209, 22])
```

```python
steag_arr[:, :, :3 ] = BACKGROUND_COLOR 
```

```python
horizontal_height = HEIGHT // 5
horizontal_start = HEIGHT // 2 - HEIGHT // 10
steag_arr[:, horizontal_start: horizontal_start+ horizontal_height , :3 ] = CROSS_COLOR
```

```python
vertical_width = horizontal_height
vertical_start = HEIGHT // 2 - vertical_width // 2
steag_arr[vertical_start: vertical_start + vertical_width, :  , :3 ] = CROSS_COLOR
```

```python
new_image = Image.fromarray(steag_arr)
new_image.save("steag_suedia.png")
```
