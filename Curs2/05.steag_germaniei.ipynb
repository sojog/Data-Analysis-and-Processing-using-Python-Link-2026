```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
BLACK = np.array([0, 0, 0])
YELLOW = np.array([255, 204, 0])
RED = np.array([221, 0, 0])
```

```python
SCALE = 1 * 100

WIDTH = 15 * SCALE
HEIGHT = 9 * SCALE
```

```python
WIDTH, HEIGHT = HEIGHT, WIDTH
```

```python
steag_arr = np.ones((HEIGHT, WIDTH, 4), dtype=np.uint8)
```

```python
steag_arr[:,:,-1] = 255
steag_arr
```

```python
new_image = Image.fromarray(steag_arr)
new_image.save("steag_negru.png")
```

```python
steag_arr[:,0:WIDTH // 3, :3 ] = BLACK 
```

```python
steag_arr[:, WIDTH // 3 : 2 * WIDTH // 3, :3 ] = RED 
```

```python
steag_arr[:,  2 * WIDTH // 3 : , :3 ] = YELLOW
```

```python
steag_arr = np.rot90(steag_arr, 1, (1,0))
```

```python
new_image = Image.fromarray(steag_arr)
new_image.save("steag_germania.png")
```
