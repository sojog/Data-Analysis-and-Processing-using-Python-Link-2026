```python
import numpy as np
```

```python
start_date = 1900
stop_date = 2100
```

```python
all_years = np.arange(start_date, stop_date + 1)
all_years
```

# Varianta 1 - conditii

```python
divizibil_cu_400 = all_years % 400 == 0
divizibil_cu_4  = all_years % 4 == 0
nedivizibil_cu_100 = all_years % 100 != 0

all_years[divizibil_cu_400 | (divizibil_cu_4 & nedivizibil_cu_100)]
```

```python
all_years[(all_years % 400 == 0) | ((all_years % 4 == 0) & (all_years % 100 != 0))]
```

# Varianta 2 - cu adunarea array-urilor

```python
divizibil_cu_400 = all_years % 400 == 0
years_400s = all_years[divizibil_cu_400]
years_400s
```

```python
divizibil_cu_4  = all_years % 4 == 0
nedivizibil_cu_100 = all_years % 100 != 0

years_4_not_100s = all_years[divizibil_cu_4 & nedivizibil_cu_100]
years_4_not_100s
```
