# Pandas  are Serie & DataFrame

# 1. Series -> este un array specific Pandas (o coloana din tabel)

```python
serie = pd.Series(np.arange(100, 106))
serie
```

```python
type(serie), serie.dtype
```

```python
serie % 2 == 0
```

```python
serie[serie % 2 == 0]
```

```python
serie + 2
```

```python
serie.max(), serie.sum()
```

# 2. DataFrame - matrice multidimensionala (un tabel cu mai multe coloane)

```python
df = pd.DataFrame(np.arange(100, 106))
df
```

```python
type(df), df.dtypes
```

```python
df = pd.DataFrame ( {"numere": np.arange(100, 106)})
df
```

```python
df = pd.DataFrame([np.arange(100, 106),np.arange(200, 206)])
df
```

```python
df = pd.DataFrame({"Coloana A": np.arange(100, 106), "Coloana B":np.arange(200, 206)})
df
```

```python
df.dtypes
```

```python
dictionar = {"culori": ["rosu", "verde", "negru", "alb"], "temperaturi":np.random.randint(0, 50, 4)}

df = pd.DataFrame(dictionar)
df
```

```python
df.dtypes
```

```python
df.min()
```

```python
df.describe()
```
