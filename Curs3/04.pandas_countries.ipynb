# 15. Filtrarea dupa valori care nu sunt nule

```python
df[~df["population"].isna()]
```

# 16. Crearea unui nou DataFrame

```python
country_population_df  = df[~df["population"].isna()][["country", "population"]]
country_population_df
```

# 17. Sortarea unui DataFrame dupa coloane

```python
country_population_df.sort_values("country")
```

```python
country_population_df.sort_values("population", ascending=False)
```

# 18. Salvarea unui fisier csv

```python
country_population_df.sort_values("country").to_csv("countries_population.csv")
```
