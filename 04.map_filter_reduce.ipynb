# 1. Map

```python
orase = ["Milano", "Bucuresti", "Barcelona", "Londra", "Madrid", "Viena", "Cluj"]
```

```python
lungime_orase = [ len(i) for i in orase ]
lungime_orase
```

```python
def functie_de_mapare(x):
    return len(x)
```

```python
list(map( functie_de_mapare , orase))
```

```python
list(map( len , orase))
```

Ex: Boltz : dacă numărul este multiplu de 7 sau contine 7 se inlocuieste cu "boltz"

```python
lista_numere = list(range(100))
print(lista_numere)
```

```python
def mapeaza_boltz(numar):
    if numar % 7 == 0 or "7" in str(numar):
        return "boltz"
    return numar
```

```python
print(list(map(mapeaza_boltz, range(1,100))))
```

# 2. Filter

```python
orase = ["Milano", "Bucuresti", "Barcelona", "Londra", "Madrid", "Viena", "Cluj"]
```

```python
# Functia de filtrare trebuie sa intoarca True (daca ramane) sau False (daca il elimina)
def filtreaza_orasul_cu_b(oras:str):
    return oras.startswith("B")
```

```python
list(filter(filtreaza_orasul_cu_b, orase))
```

```python
def este_lungimea_mai_mica_decat_6(oras:str):
    return len(oras) < 6
```

```python
list(filter(este_lungimea_mai_mica_decat_6, orase))
```

```python
list(filter(lambda x: x.endswith("a"), orase))
```

# 3. Reduce

```python
from functools import reduce
```

```python
lista = [2, 3, 4, 5]
```

```python
def suma_numerelor(x, y):
    print("x=", x, "y=", y)
    return x + y
```

```python
reduce(suma_numerelor, lista)
```

```python
def concatenare(x, y):
    print("x=", x, "y=", y)
    return str(x) + str(y)
```

```python
reduce(concatenare, lista)
```

```python
reduce(lambda x, y: str(x)+str(y), lista)
```

```python
lista = [2, 3, 4, 5]
reduce(lambda x, y: x if x > y else y, lista)
```
