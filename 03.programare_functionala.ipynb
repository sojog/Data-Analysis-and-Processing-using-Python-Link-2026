# Functiile pot fi pasate ca parametru

```python
def adunare(x, y):
    return x + y
```

```python
def scadere(x, y):
    return x - y
```

```python
def inmultire(x,y):
    return x * y
```

```python
for op in [adunare, scadere, inmultire]:
    print("Rezultatul", op, "este:", op(10, 2))
```

```python
def executa_operatia(operatie, termen1, termen2):
    return operatie(termen1, termen2)
```

```python
executa_operatia(adunare, 2, 3)
```

```python
executa_operatia(inmultire, 20, 4)
```
